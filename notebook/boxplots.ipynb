# Read data from CSV

```python
import sys
from pathlib import Path
sys.path.append(str(Path("..")))
from scripts.read_data import read_data

data = read_data()
data.head()
```

# Admission grade vs gender

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(data, x = "Gender", y ="Admission grade", hue="Target")

```

* no visible correlation between gender and admission grade
* no visible difference between genders

# Admission grade vs Course

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data, x = "Course", y ="Admission grade", )
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

* no visible correlation between admission grade and course

# Admission grade vs Nacionality

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data, x = "Nacionality", y ="Admission grade")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

* no visible correlation between admission grade and nacionality

# Admission grade vs age

```python
plt.figure(figsize=(14, 6))
sns.boxplot(data, x = "Age at enrollment", y ="Admission grade")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

* we can see that as the age of students at enrollment increases, their admission grades become more varied.

# Admission grade vs status

```python
plt.figure(figsize=(9, 6))
sns.boxplot(data, x = "Target", y ="Admission grade", hue="Scholarship holder")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

* no correlation between students status (dropout, graduate and enroller) and the admission score, which seems weird to me

# Age at enrollment vs status

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data, x = "Target", y ="Age at enrollment", hue="Gender")
plt.tight_layout()
```

* correlation between the decision of dropping out and age at enrollement - the older students starts studying, there is more chance that they will not finish

* men are more willing to start studying at older age

# Age at enrollment vs nacionality

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data, x = "Nacionality", y ="Age at enrollment")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

* Many elderly or older students are from Portugal
* Brazilian, Spanish, Angolan and Moldovan students tend to enroll later
