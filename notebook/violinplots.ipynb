# Admisssion grade vs target

# Curricular units 1st sem (evaluations) vs Target

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data, x = "Target", y = "Curricular units 1st sem (evaluations)", hue = "Scholarship holder")
```

More droupouts had worse evaluations in the first semester than graduates and enrolled

Moreover dropouts and enrolled students are more likely to have really good evaluations compared to graduates

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data, x = "Target", y = "Curricular units 1st sem (evaluations)", hue = "Debtor")
```

Dropouts with debt are less likely to have bad evaluations compared to dropouts without debt

# Curricular units 1st sem (approved) vs Target

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data, x = "Target", y = "Curricular units 1st sem (approved)", hue = "Scholarship holder")
```

* Most dropouts fail many classes in the 1st semester
* Dropouts without schoolarships fail more classes than dropouts with schoolarship
