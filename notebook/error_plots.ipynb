# Read data from CSV

```python
import sys, os 
from pathlib import Path
sys.path.append(str(Path("..")))
from scripts.read_data import read_data

os.makedirs("../generated_charts", exist_ok = True)

data = read_data()
data.head()
```

```python
from scripts.errorbar import errorbar
import matplotlib.pyplot as plt

TYPE = "errorplot"
```

```python
errorbar(data["Age at enrollment"])
plt.savefig(f"../generated_charts/{TYPE}_age.png", dpi = 300, bbox_inches = "tight")
plt.show()
```

```python
errorbar(data["Previous qualification (grade)"])
plt.savefig(f"../generated_charts/{TYPE}_previous_qualification_grade.png", dpi = 300, bbox_inches = "tight")
plt.show()

errorbar(data[data["Target"] == "Dropout"]["Previous qualification (grade)"], addition = "(only dropouts)")
plt.savefig(f"../generated_charts/{TYPE}_previous_qualification_grade_dropouts.png", dpi = 300, bbox_inches = "tight")
plt.show()
```

On average, dropouts score a few points lower on the previous qualification grade.

```python
errorbar(data["Admission grade"])
plt.savefig(f"../generated_charts/{TYPE}_admission_grade.png", dpi = 300, bbox_inches = "tight")
plt.show()

errorbar(data[data["Target"] == "Dropout"]["Admission grade"], addition = "(only dropouts)")
plt.savefig(f"../generated_charts/{TYPE}_admission_grade_dropouts.png", dpi = 300, bbox_inches = "tight")
plt.show()
```

On average, dropouts score a few points lower on the admission grade.
