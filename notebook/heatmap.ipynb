# Read data from CSV

```python
import sys, os
from pathlib import Path
sys.path.append(str(Path("..")))
from scripts.read_data import read_data

os.makedirs("../generated_charts", exist_ok = True)

data = read_data()
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = data.corr(numeric_only=True)

plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, vmin = -1, annot = True, cmap="coolwarm", fmt=".2f", linewidths = .4)
plt.title("Heatmap")
plt.savefig("../generated_charts/heatmap.png", dpi = 300, bbox_inches = "tight")
plt.show()
```

* No surprise that curricular units 1st sem params (credited, enrolled, evaluations, approved, grade) are correlated
* High correlation between admission grade 
* Small negative correlation between age at enrollment and grades
* Negative correlation between GDP and unemployment rate
* Inflation rate isn't significantly correlated with GDP and unemployment
